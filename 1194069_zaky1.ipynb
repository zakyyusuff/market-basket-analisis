```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv(r'C:\Users\AKU\tugas_prog\bu nisa\market basket analisis\Market_Basket_Optimisation.csv',header=None)
```

```python
df.head()
```

```python
#untuk menggunakan aprori perlu mengonversi data dalam format daftar..

# transaksi = [['apple','almond'],['apple'],['banana','apple']]...
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
transactions[0]
```

```python
#Panggil fungsi apriori yang membutuhkan dukungan, kepercayaan, dan angkat minimum, panjang minimum adalah kombinasi dari item default adalah 2".
rules = apriori(transactions,min_support=0.003,min_confidance=0.2,min_lift=3,min_length=2)
#itu menghasilkan seperangkat aturan dalam file generator...
rules
```

```python
# semua aturan perlu dikonversi dalam daftar..
Results = list(rules)
Results
```

```python
#convert hasil dalam kerangka data untuk operasi lebih lanjut...
# seperti yang kita lihat statistik pesanan itu sendiri daftar jadi perlu dikonversi dalam format yang tepat..
df_results = pd.DataFrame(Results)
df_results.head()
```
